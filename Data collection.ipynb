```python
driver = webdriver.Chrome('C:/Users/maorr/Downloads/chromedriver.exe')
driver.get('https://www.goodreads.com/list/show/2451.I_Saw_the_Movie_Read_the_Book?page=4')
time.sleep(2)
```

```python
### שמירת הליניקים של הדף הנוכחי
links=[]
elems = driver.find_elements_by_xpath("//a[@href][@class='bookTitle']")
for elem in elems:
    links.append(elem.get_attribute("href"))


current_url ='https://www.goodreads.com/list/show/2451.I_Saw_the_Movie_Read_the_Book?page=4'
```

```python
k=0
for l in range(1,1500):
        driver.get(links[k])
        time.sleep(5)  
        
         ### סגירת גרסת בטא
        try :
            driver.find_element_by_xpath("//button[@class='Button Button--transparent Button--small Button--rounded']")
            driver.get(links[k])
            time.sleep(5)  
            window_name = driver.window_handles[0]
            driver.switch_to.window(window_name=window_name)
            driver.close()
            driver.switch_to.window(driver.window_handles[1])
        except :
            pass
        
        
        
        ### סגירת מודעה
        try :
            driver.find_element_by_xpath("//img[@src='//s.gr-assets.com/assets/gr/icons/icon_close-63734f04e7baaa77fbad796225e5724c.svg']/parent::button").click()
        except:
            pass
        
        
                 
    
        ### more deitals כפתור
        try:
            driver.find_element_by_xpath("//a[@class = 'left inter actionLinkLite']").click()
            
        except:
            pass
        
        ###  משיכת נתונים גודרידס
        try :
            head = driver.find_element_by_xpath("//h1[@class = 'gr-h1 gr-h1--serif']").text
            Title.append(head)
        except :
            Title.append(np.nan)
            
        try :
            author = driver.find_element_by_xpath("//a[@class = 'authorName']").text
            Name_of_author.append(author)
        except :
            Name_of_author.append(np.nan)
            
        try :
            rating = float(driver.find_element_by_xpath(f"//span[@itemprop='ratingValue']").text)
            Rating.append(rating)
        except :
            Rating.append(np.nan)
        try :
            rating_counter = int(driver.find_element_by_xpath("//*[@itemprop='ratingCount']").get_attribute("content"))
            Rating_count.append(rating_counter)
        except :
            Rating_count.append(np.nan)
            
        try :
            review_counter = int(driver.find_element_by_xpath("//*[@itemprop='reviewCount']").get_attribute("content"))
            Review_count.append(review_counter)
        except :
            Review_count.append(np.nan)
            
        try :
            cover = driver.find_element_by_xpath(f"//span[@itemprop='bookFormat']").text
            Book_cover.append(cover)
        except:
            Book_cover.append(np.nan)
            
        try :
            Num_of_editions = driver.find_element_by_xpath("//*[@class='otherEditionsLink']").text
            Num_of_editions=Num_of_editions.lstrip(" Other Editions (").rstrip(" ) ")
            Num_of_other_editions.append(Num_of_editions)
        except :
            Num_of_other_editions.append(np.nan)
        
        try :
            geners= driver.find_element_by_xpath(f"//div[@class='left']")
            Genre.append(geners.text)
        except :
            Genre.append(np.nan)
        
        ### more awards 
        try :
            driver.find_element_by_xpath("//span[@class = 'toggleLink']").click()
        except :
            pass
                
        
        counter_awards =0
        try:
            awards = driver.find_element_by_xpath(f"//div[@itemprop='awards']").text
            if awards :
                counter_awards =1
                for i in awards:
                    if i == ',':
                        counter_awards+=1
                    
                Num_of_awards.append(counter_awards)
        
        except: 
            Num_of_awards.append(counter_awards)
            
            
            
        try:
            pages=driver.find_element_by_xpath("//span[@itemprop='numberOfPages']").text
            Num_of_pages.append(int(pages.rstrip(" pages ")))
        except:
            Num_of_pages.append(np.nan)
        
        
        
        
        try:
            x=driver.find_elements_by_xpath('//*[@class="row"]')[1].text.split(" by ")
            year=x[0][-4:]
            publisher=x[1]
            try:
                publisher=publisher.split("(")[0]
            except:
                pass
            try:
                publisher=publisher.rstrip(" . ")
            except:
                pass
            Publisher.append(publisher)
            Release_date.append(int(year))
        except:
            Publisher.append(np.nan)
            Release_date.append(np.nan)
        
        
        
        
        try:
            language=driver.find_element_by_xpath('//*[@itemprop="inLanguage"]').text
            Edition_language.append(language)
        except:
            Edition_language.append(np.nan)
        
            
        is_movie.append(1)
        time.sleep(2)

    ### ---------------------------------------------
        ## פתיחת דף של הספר בנובל
        try :
            action=webdriver.ActionChains(driver)
            action.move_to_element(driver.find_element_by_xpath("//*[@id='buyDropButtonStores']"))
            action.perform()
            time.sleep(5)
            driver.find_element_by_partial_link_text("Barnes & Noble").click()
            time.sleep(5)
            driver.switch_to.window(driver.window_handles[1])
            time.sleep(5)
            try:
                for li in (driver.find_elements_by_xpath("//a[@class=' link']//span")):
                    if(li.text)=='Paperback':
                        li.click()
                        break
            except:
                pass
        except :
            print("nope")
            Series.append(np.nan)
            min_Reading_age.append(np.nan)
            Lexile_measure.append(np.nan)
            best_seller.append(np.nan)
            
            
        ###משיכת נתונים-נובל
        
        
        book_titles=[]#כותרות

        try:
            elems=driver.find_elements_by_xpath("//table[@class='plain centered']//th")
            for elem in elems:
                book_titles.append(elem.text.rstrip(' : '))
        except:
               pass

            
        book_info=[]#פרטים
        try:
            elems=driver.find_elements_by_xpath("//table[@class='plain centered']//td")
            for elem in elems:
                book_info.append(elem.text)
        except:
            pass

        
        for i in range(len(book_titles)):
            if(book_titles[i]=='Age Range'):
                try:
                    min_Reading_age.append(book_info[i].rstrip(" Years "))
                except:
                    pass
            elif(book_titles[i]=='Lexile'):
                try:
                    book_info[i]=book_info[i].rstrip(" L (what's this?)")
                    try:
                        book_info[i]=book_info[i].lstrip("HL")
                    except:
                        pass
                    Lexile_measure.append(book_info[i])
                except:
                    pass
                
            elif(book_titles[i]=='Sales rank'):
                try:
                    best_seller.append(int(book_info[i].replace(',', '')))
                except:
                    try:
                        best_seller.append(int(book_info[i]))
                    except:
                        pass
                    
            elif (book_titles[i]=='Series'):
                Series.append(1)
                


        for i in range(1):
            if (len(Num_of_pages)==l-1):
                Num_of_pages.append(np.nan)
            if (len(Edition_language)==l-1):
                Edition_language.append(np.nan)
            if (len(Publisher)==l-1):
                Publisher.append(np.nan)
            if (len(Release_date)==l-1):
                Release_date.append(np.nan)
            if (len(Grade_level)==l-1):
                Grade_level.append(np.nan)
            if (len(min_Reading_age)==l-1):
                min_Reading_age.append(np.nan)
            if (len(Lexile_measure)==l-1):
                Lexile_measure.append(np.nan)
            if (len(Series)==l-1):
                Series.append(0)
            if(len(best_seller)==l-1):
                best_seller.append(np.nan)
                

        
        
        ### לסגור את הדף של נובל
        try:
            window_name = driver.window_handles[1]
            driver.switch_to.window(window_name=window_name)
            driver.close()
        except:
            pass
        
        
        window_name = driver.window_handles[0]
        driver.switch_to.window(window_name=window_name)
        k+=1
        ###  מעבר לעמוד הבא בגוד רידס
        if (k%99 == 0 and k != 0):
            driver.get(current_url)
            driver.find_element_by_class_name('next_page').click()
            current_url= driver.current_url
            k=0
            links=[]
            elems = driver.find_elements_by_xpath("//a[@href][@class='bookTitle']")
            for elem in elems:
                links.append(elem.get_attribute("href"))
                time.sleep(0.1)
```
