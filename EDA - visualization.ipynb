```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt      
import math
import seaborn as sns
from collections import Counter
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
df=pd.read_csv('Data set after cleaning.csv')
df['Rating count']=df['Rating count'].astype(float)
df['Num of awards']=df['Num of awards'].astype(float)
df['Edition language']=df['Edition language'].astype(float)
df['Reading age']=df['Reading age'].astype(float)
df
```

```python

sr = df['There is a movie'].value_counts()[:2]
sr.plot(kind='pie',labels=['Yes','No'],autopct='%1.1f%%')
plt.title('Books made into movies')
```

```python
plt.scatter(df['Num of awards'], df['Review count'])
plt.xlabel('Num of awards')
plt.ylabel('Review count')
plt.show()
df['Num of awards'].corr(df['Review count'])
```

```python
df['Release date'].hist(bins=10, grid = False, rwidth=0.9, color ='blue')
plt.xlabel('Release date')
plt.ylabel('Amount')
plt.title('Year')
plt.show()
df['There is a movie'].corr(df['Release date'])
```

```python
from mpl_toolkits.mplot3d import Axes3D
ax=plt.axes(projection='3d')
x = df['Rating count']
y= df['Review count']
z=df['Rating']

plt.xlabel('Rating count')
plt.ylabel('Review count')
ax.scatter3D(x,y,z, c=z, depthshade=False)
```

```python
ct = pd.crosstab(df['Reading age'] , df['There is a movie'], normalize ='index')
ct.plot.bar(stacked=True)
plt.show()
```
